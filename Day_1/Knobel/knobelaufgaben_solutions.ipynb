# Challenge 4) If statements and Loops

Because Python's birthday is coming up I want you to bake it its favourite cake! The ingredients you need for it were given to you after completing the previous challenge and should be assigned to the variable `ingredients`. I made another list of what you already have at home (I've been watching you) called `have`.

I need you to find out which ingredients you need to buy at the store. To achieve this, loop through the `ingredients` list and check if each item is also in the `have` list. Only add the item to the `buy` list if it is not already in the `have` list.

<div class="alert alert-block alert-success">
<b>Solution:</b> The basic idea is the following: we loop through all elements of the <code>ingredients</code> list using a for-loop. In each iteration (in other words: for each ingredient) we check whether the current ingredient is NOT contained in the <code>have</code> list using the expression <code>if (not ingredient in have): ...</code>. This expression is made up of two main parts: <code>in</code> and <code>not</code>. The first of the two, <code>in</code> lets us check, whether <code>ingredient</code> is contained in <code>have</code>. That is, the expression <code>ingredient in have</code> returns the value <code>True</code> if the current ingredient is contained in the <code>have</code>-list. Since we only care about the cases when an ingredient is not contained in the <code>have</code>-list, we use <code>not</code> to invert the condition. If some ingredient is NOT contained in the <code>have</code>-list, we add it to the <code>buy</code>-list.
</div>

````python
have = ["Banana","Sugar"]
buy = []

# PUT YOUR CODE BELOW

# loop through all ingredients which are given in the ingredients list
for ingredient in ingredients:
    # check whether the current ingredient is NOT on the 'have' list
    if not ingredient in have:
        # if the current ingredient is not on the 'have' list, add it to the 'buy' list
        buy.append(ingredient)

# PUT YOUR CODE ABOVE

print(buy)  # prints your solution
````

# Challenge 6) Crack the code

As your final challenge, I have hidden your Python behind one of these doors (see image below). To find the correct door you need to crack the mysterious code you got after completing the previous task.

Here are instructions on how to interpret the code:
```
'*'  : move one door to the right

'#'  : move one door to the left

'->' : move three doors to the right

'<-' : move three dors to the left
```

**Important:** You start with door 0. 

![image](doors.png)

Your task is to programm a function, that takes the code as an input and returns the door that Python is hiding behind. Go through each symbol and solve the riddle!

<div class="alert alert-block alert-success">
<b>Solution:</b> This function is a tough one. It combines many concepts we learned today: functions, loops and flow control. Our function takes a variable <code>code</code> which is supposed to be a list of symbols as the one given by the challenger, and it returns the correct door number according to the mapping given in the challenge description. Initially we set the current door to 0. This is given in the task description as well.

We then loop through all symbols of the code and check how many doors we need to move in which direction. We increase or decrease the <code>current_door</code> variable accordingly and in the end we return the final value of <code>current_door</code>, which is the one behind which Python is hidden.
</div>

````python
def find_door(code):
    
    current_door=0
    
    # PUT YOUR CODE BELOW
    
    # loop through all symbols
    for symbol in code:
        # check whether we need to move one door to the right
        if symbol == "*":
            current_door = current_door + 1
        # check whether we need to move one door to the left
        elif symbol == "#":
            current_door = current_door - 1
        # check whether we need to move three doors to the right
        elif symbol == "->":
            current_door = current_door + 3
        # check whether we need to move three doors to the left
        elif symbol == "<-":
            current_door = current_door - 3
        
    # PUT YOUR CODE ABOVE
    
    return current_door


solution = find_door(mysterious_code)
print(solution)  # prints out your solution
````
